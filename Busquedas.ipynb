#Descripción de los algoritmos

##búsqueda secuencial

AQUI LAS DESCRIPCCIONES:
1. Comenzamos desde el principio de la lista.
2. Miramos el primer elemento y lo comparamos con lo que estamos buscando.
3. Si coincide, entonces encontramos lo buscado.
4. Si no coincide, avanzamos al siguiente elemento de la lista.
5. Repetimos los pasos 2 a 4 hasta encontrar lo que buscamos o hasta revisar toda la lista.

##búsqueda binaria

AQUI LAS DESCRIPCCIONES:
1. Comenzamos con una lista ordenada y definimos dos puntos, uno al principio y otro al final.
2. Calculamos el punto medio entre esos dos puntos.
3. Comparamos el elemento en el punto medio con lo que estamos buscando.
4. Si coincide, entonces encontramos lo buscado.
5. Si el punto medio es menor que lo que buscamos, avanzamos nuestro punto de inicio al punto medio + 1.
6. Si el punto medio es mayor que lo que buscamos, movemos nuestro punto final al punto medio - 1.
7. Repetimos los pasos 2 a 6 hasta encontrar lo que buscamos o hasta que los puntos de inicio y final se crucen.

#Análisis de los algoritmos

##Busqueda Secuencial Iterativa Clasificacion (Memoria y Tiempo)

1. **Búsqueda Secuencial Iterativa (Tiempo)**:
   - Clasificación: O(n)
   - Definición: El tiempo de ejecución crece linealmente con el tamaño de entrada, ya que en el peor caso, el algoritmo debe recorrer todos los elementos uno por uno hasta encontrar el elemento deseado.

   **No es necesario una demostracion tan formal ya que este se puede notar con el siguiente ejemplo el cual se basa en el peor caso:**

    T(1) = 1

    T(2) = 2

    T(3) = 3

    T(4) = 4

    .
    .......
    .

    T(n) = n

2. **Búsqueda Secuencial Iterativa (Espacio)**:
   - Clasificación: O(1)
   - Definición: El algoritmo utiliza una cantidad constante de memoria independientemente del tamaño de entrada, ya que no necesita almacenar estructuras de datos adicionales.

##Busqueda Secuencial Recursiva Clasificacion (Memoria y Tiempo)
1. **Búsqueda Secuencial Recursiva (Tiempo)**:
   - Clasificación: O(n)
   - Definición: Al igual que la versión iterativa, en el peor caso, este algoritmo también debe recorrer todos los elementos uno por uno, lo que resulta en un tiempo de ejecución lineal.

   **La demostracion es analaloga a la vercion iterativa**

2. **Búsqueda Secuencial Recursiva (Espacio)**:
   - Clasificación: O(n)
   - Definición: El algoritmo utiliza espacio en la pila de llamadas para cada llamada recursiva, lo que puede alcanzar una profundidad de n en el peor caso (si el elemento no se encuentra).

##Busqueda Binaria Iterativa clasificacion (Memoria y Tiempo)

1. **Búsqueda Binaria Iterativa (Tiempo)**:
   - Clasificación: O(log n)
   - Definición: El tiempo de ejecución crece de manera logarítmica con el tamaño de entrada, ya que el algoritmo divide el espacio de búsqueda a la mitad en cada iteración.

  **LA DEMOSTRACION DE ESTE ES COMPLETAMENTE ANALOGA ALA VERCIÓN RECURSIVA EXPLICADA MAS ADELANTE**

2. **Búsqueda Binaria Iterativa (Espacio)**:
   - Clasificación: O(1)
   - Definición: Al igual que la versión iterativa de la búsqueda secuencial, este algoritmo utiliza una cantidad constante de memoria independientemente del tamaño de entrada.

##Busqueda Binaria Recursiva clasificacion (Memoria y Tiempo)

1. **Búsqueda Binaria Recursiva (Tiempo)**:
   - Clasificación: O(log n)
   - Definición: Al igual que la versión iterativa, el tiempo de ejecución de la búsqueda binaria recursiva crece de manera logarítmica con el tamaño de entrada debido a la división recursiva del espacio de búsqueda.

   $$T(n)=\begin{cases}
b, & n=0\,\acute{o}\,n=1\\
T(\frac{n}{2})+a & n\geq2
\end{cases}$$


Sea $n = 2^k$



$$T(n)  =  T(2^{k-1})+a$$

Repitiendo por definicion

$$T(n)  = (T(2^{k-2})+a)+a$$

Notemos que nos interesa llegar al paso base donde $T(1) = b$ por lo que si nos fijamos detenidamente en los exponentes llegamos a la conclusion de que para que esto pase tenemos que tener que  $b = T(1) = T(2^{0}) = T(2^{k-k})$.
Ahora para hacer que esto suceda notemos que el algoritmo cada vez que lo ejecutamos se baja en 1 el exponente por lo que si ejecutamos el algoritmo $k$ veces llegaremos al paso base a si que hagamos esto ultimo:

LLegamos a donde queriamos ahora solo es cuestion de un poco de algebra y dirigirnos hacia la definicion para tener que:

Ahora para encontrar la cota y el orden de complejidad notemos que tanto $a$ como $b$ son constantes por lo qur fijemosnos que la parte que crece mas rapido es:

#Código de los algoritmos de busqueda

```python
#Búsqueda Secuencial Iterativa
def b_secuencial(coleccion, valor):
  # Recorremos la colección utilizando un bucle for y un índice i que varía de 0 a la longitud de la colección - 1
  for i in range(len(coleccion)):
    # Comparamos el elemento actual de la colección con el valor que estamos buscando
    if coleccion[i] == valor:
      # Si encontramos una coincidencia, retornamos el índice en el que se encontró el valor
      return i
  # Si recorremos toda la colección y no encontramos el valor, retornamos False para indicar que no se encontró
  return False
```

```python
# Definición de la función de búsqueda secuencial recursiva
def busqueda_secuencial_recursiva(lista, elemento, indice=0):
    # Verificamos si el índice es igual o mayor que la longitud de la lista.
    if indice >= len(lista):
        # Si el índice es mayor o igual a la longitud de la lista, significa que hemos llegado al final de la lista
        # y no hemos encontrado el elemento, por lo que retornamos -1 para indicar que el elemento no se encontró.
        return -1
    # Comparamos el elemento actual de la lista con el elemento que estamos buscando.
    if lista[indice] == elemento:
        # Si encontramos una coincidencia, retornamos el índice en el que se encontró el elemento.
        return indice
    # Si no encontramos una coincidencia en el índice actual, llamamos recursivamente a la función con el siguiente índice.
    return busqueda_secuencial_recursiva(lista, elemento, indice + 1)


```

```python
#Búsqueda Binaria Recursiva
def busquedaBinariaR(lista, dato, inicio, final):
  # Caso base: Si el índice final es menor que el índice de inicio, significa que no encontramos el elemento en la lista.
  if final < inicio:
    return False

  # Calculamos el índice de la mitad de la lista
  mitad = (inicio + final) // 2

  # Caso base: Si el elemento en la posición de la mitad es igual al dato que estamos buscando, retornamos la posición de la mitad.
  if lista[mitad] == dato:
    return mitad

  # Si el dato es menor que el elemento en la posición de la mitad, hacemos una llamada recursiva
  # a la función en la mitad izquierda de la lista (desde inicio hasta mitad-1).
  if dato < lista[mitad]:
    return busquedaBinariaR(lista, dato, inicio, mitad-1)
  else:
    # Si el dato es mayor que el elemento en la posición de la mitad, hacemos una llamada recursiva
    # a la función en la mitad derecha de la lista (desde mitad+1 hasta final).
    return busquedaBinariaR(lista, dato, mitad+1, final)
```

```python
#Búsqueda Binaria Iterativa
def busqueda_binaria_iterativa(lista, elemento):
    # Establecemos dos punteros, izquierda al principio de la lista y derecha al final de la lista.
    izquierda, derecha = 0, len(lista) - 1

    # Mientras el puntero izquierdo sea menor o igual al puntero derecho, continuamos buscando.
    while izquierda <= derecha:
        # Calculamos el índice del elemento medio.
        medio = (izquierda + derecha) // 2

        # Comparamos el elemento medio con el elemento que estamos buscando.
        if lista[medio] == elemento:
            # Si encontramos una coincidencia, retornamos el índice del elemento.
            return medio
        elif lista[medio] < elemento:
            # Si el elemento medio es menor que el elemento buscado,
            # restringimos la búsqueda a la mitad derecha de la lista.
            izquierda = medio + 1
        else:
            # Si el elemento medio es mayor que el elemento buscado,
            # restringimos la búsqueda a la mitad izquierda de la lista.
            derecha = medio - 1

    # Si llegamos aquí, significa que no se encontró el elemento en la lista,
    # por lo que retornamos -1 para indicar que no se encontró.
    return -1
```

```python
import time
import random
import matplotlib.pyplot as plt

# Algoritmo de ordenamiento: Insertion Sort
def insertionSort(lista):
    for i in range(1, len(lista)):
        key = lista[i]
        j = i - 1
        while j >= 0 and key < lista[j]:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = key

# Función para medir el tiempo de ejecución de un algoritmo de ordenamiento
def medir_tiempo_ordenamiento(algoritmo, lista):
    inicio = time.time()
    algoritmo(lista.copy())  # Copiamos la lista para no modificar la original
    fin = time.time()
    return fin - inicio

# Tamaños de las colecciones de datos
tamaños = [10, 100, 1000, 10000]

# Diccionario para almacenar los tiempos de cada algoritmo
tiempos = {"Insertion Sort": []}

for tamaño in tamaños:
    # Generar una lista aleatoria de tamaño 'tamaño'
    lista_aleatoria = random.sample(range(1, tamaño * 10), tamaño)

    # Medir el tiempo de ordenamiento para Insertion Sort
    tiempo = medir_tiempo_ordenamiento(insertionSort, lista_aleatoria)
    tiempos["Insertion Sort"].append(tiempo)

# Generar la gráfica
for algoritmo, tiempo_lista in tiempos.items():
    plt.plot(tamaños, tiempo_lista, label=algoritmo)

plt.xlabel("Número de Datos")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs. Número de Datos para Algoritmos de Ordenamiento")
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Algoritmo de ordenamiento: Insertion Sort
def SelectionSort(lista):
    # La función "SelectionSort" toma una lista como entrada para ser ordenada.

    for i in range(len(lista)):
        # El primer bucle "for" itera a través de cada elemento en la lista.

        minimo = i
        # Inicializamos la variable "minimo" con el índice actual "i" como el índice del elemento mínimo.

        for j in range(i + 1, len(lista)):
            # El segundo bucle "for" itera desde el índice siguiente al actual "i" hasta el final de la lista.

            if lista[j] < lista[minimo]:
                # Comparamos el elemento actual con el elemento mínimo encontrado hasta ahora.

                # Si encontramos un elemento menor, actualizamos el índice del elemento mínimo.
                minimo = j

        # Después de completar el segundo bucle, hemos encontrado el elemento mínimo en la parte no ordenada de la lista.
        # Ahora intercambiamos el elemento mínimo con el elemento en la posición actual "i".
        lista[i], lista[minimo] = lista[minimo], lista[i]

    # Después de que ambos bucles hayan terminado, la lista estará ordenada en orden ascendente.
    return lista  # Devolvemos la lista ordenada

# Función para medir el tiempo de ejecución de un algoritmo de ordenamiento
def medir_tiempo_ordenamiento(algoritmo, lista):
    inicio = time.time()
    algoritmo(lista.copy())  # Copiamos la lista para no modificar la original
    fin = time.time()
    return fin - inicio

# Tamaños de las colecciones de datos
tamaños = [10, 100, 1000, 10000]

# Diccionario para almacenar los tiempos de cada algoritmo
tiempos = {"SelectionSort Sort": []}

for tamaño in tamaños:
    # Generar una lista aleatoria de tamaño 'tamaño'
    lista_aleatoria = random.sample(range(1, tamaño * 10), tamaño)

    # Medir el tiempo de ordenamiento para Insertion Sort
    tiempo = medir_tiempo_ordenamiento(SelectionSort, lista_aleatoria)
    tiempos["SelectionSort Sort"].append(tiempo)

# Generar la gráfica
for algoritmo, tiempo_lista in tiempos.items():
    plt.plot(tamaños, tiempo_lista, label=algoritmo)

plt.xlabel("Número de Datos")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs. Número de Datos para Algoritmos de Ordenamiento")
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

def bubbleSort(lista):
    # La función "bubbleSort" toma una lista como entrada para ser ordenada.

    n = len(lista)
    # Obtenemos la longitud de la lista para conocer cuántos elementos hay en ella.

    intercambio = True
    # Inicializamos la variable "intercambio" como True para entrar en el primer bucle while.

    j = 0
    # Inicializamos una variable "j" que se usará para optimizar el rendimiento del algoritmo.

    while intercambio:
        # Iniciamos un bucle "while" que continuará hasta que no se realicen más intercambios.

        j = j + 1
        # Incrementamos "j" en cada iteración, ya que sabemos que los últimos "j" elementos ya están en su lugar.

        intercambio = False
        # Suponemos que no habrá más intercambios hasta que se demuestre lo contrario.

        for i in range(n - j):
            # El bucle "for" itera a través de los elementos de la lista hasta el penúltimo elemento.

            if lista[i] > lista[i + 1]:
                # Comparamos el elemento actual con el siguiente elemento.

                # Si el elemento actual es mayor que el siguiente, los intercambiamos.
                lista[i], lista[i + 1] = lista[i + 1], lista[i]
                intercambio = True
                # Indicamos que se realizó un intercambio en esta iteración.

    # Después de que ambos bucles hayan terminado, la lista estará ordenada en orden ascendente.
    return lista

# Función para medir el tiempo de ejecución de un algoritmo de ordenamiento
def medir_tiempo_ordenamiento(algoritmo, lista):
    inicio = time.time()
    algoritmo(lista.copy())  # Copiamos la lista para no modificar la original
    fin = time.time()
    return fin - inicio

# Tamaños de las colecciones de datos
tamaños = [10, 100, 1000, 10000]

# Diccionario para almacenar los tiempos de cada algoritmo
tiempos = {"bubbleSort Sort": []}

for tamaño in tamaños:
    # Generar una lista aleatoria de tamaño 'tamaño'
    lista_aleatoria = random.sample(range(1, tamaño * 10), tamaño)

    # Medir el tiempo de ordenamiento para Insertion Sort
    tiempo = medir_tiempo_ordenamiento(bubbleSort, lista_aleatoria)
    tiempos["bubbleSort Sort"].append(tiempo)

# Generar la gráfica
for algoritmo, tiempo_lista in tiempos.items():
    plt.plot(tamaños, tiempo_lista, label=algoritmo)

plt.xlabel("Número de Datos")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs. Número de Datos para Algoritmos de Ordenamiento")
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

def MergeSort(List):
    # La función "MergeSort" toma una lista como entrada para ser ordenada.

    Sorted_list = []
    # Creamos una lista vacía "Sorted_list" que se usará para almacenar la lista ordenada.

    if len(List) <= 1:
        return List
    # Si la longitud de la lista es 1, significa que ya está ordenada, por lo que la devolvemos tal como está.

    mitad = len(List) // 2
    # Dividimos la lista en dos mitades aproximadamente iguales usando división entera.

    izquierda = MergeSort(List[:mitad])
    # Llamamos recursivamente a MergeSort en la mitad izquierda de la lista.

    derecha = MergeSort(List[mitad:])
    # Llamamos recursivamente a MergeSort en la mitad derecha de la lista.

    izquierda_len = len(izquierda)
    derecha_len = len(derecha)
    # Obtenemos las longitudes de las listas izquierda y derecha.

    i = 0
    j = 0
    # Inicializamos dos índices i y j para recorrer las listas izquierda y derecha respectivamente.

    while i != izquierda_len or j != derecha_len:
        # Iniciamos un bucle "while" que continuará hasta que hayamos procesado todos los elementos de ambas listas.

        if i != izquierda_len and (j == derecha_len or izquierda[i] < derecha[j]):
            # Comparamos el elemento actual de la lista izquierda con el elemento actual de la lista derecha.

            # Si el elemento de la lista izquierda es menor (o si ya hemos recorrido toda la lista derecha),
            # agregamos el elemento de la lista izquierda a la lista "Sorted_list" y avanzamos en la lista izquierda.
            Sorted_list.append(izquierda[i])
            i += 1
        else:
            # En caso contrario, agregamos el elemento de la lista derecha a la lista "Sorted_list" y avanzamos en la lista derecha.
            Sorted_list.append(derecha[j])
            j += 1

    # Después de que el bucle "while" haya terminado, la lista "Sorted_list" contendrá todos los elementos ordenados de las listas izquierda y derecha.
    return Sorted_list

# Función para medir el tiempo de ejecución de un algoritmo de ordenamiento
def medir_tiempo_ordenamiento(algoritmo, lista):
    inicio = time.time()
    algoritmo(lista.copy())  # Copiamos la lista para no modificar la original
    fin = time.time()
    return fin - inicio

# Tamaños de las colecciones de datos
tamaños = [10, 100, 1000, 10000]

# Diccionario para almacenar los tiempos de cada algoritmo
tiempos = {"MergeSort Sort": []}

for tamaño in tamaños:
    # Generar una lista aleatoria de tamaño 'tamaño'
    lista_aleatoria = random.sample(range(1, tamaño * 10), tamaño)

    # Medir el tiempo de ordenamiento para Insertion Sort
    tiempo = medir_tiempo_ordenamiento(MergeSort, lista_aleatoria)
    tiempos["MergeSort Sort"].append(tiempo)

# Generar la gráfica
for algoritmo, tiempo_lista in tiempos.items():
    plt.plot(tamaños, tiempo_lista, label=algoritmo)

plt.xlabel("Número de Datos")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs. Número de Datos para Algoritmos de Ordenamiento")
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Definicion recursiva de quicksort
def quickSort(lista):
  # si la lista es de tamaño 1 o menor se devuelve
  if len(lista) < 2:
    return lista
  # en otro caso se parte la lista y se resuelve recursivamente
  else:
    # se parte la lista original (n operaciones)
    menores, pivote, mayores = particion(lista)
    # 2*T(n/2)
    return quickSort(menores)+[pivote]+quickSort(mayores)

# Definicion del algoritmo para partir una lista en mayores y menores
def particion(lista):
  mayores, menores = [], []
  # pivote para realizar la comparación
  pivote = lista[0]
  for i in range(1,len(lista)):
    if lista[i] < pivote:
      menores.append(lista[i])
    else:
      mayores.append(lista[i])
  # se devuelven ambas listas y el pivote
  return menores, pivote, mayores
# Función para medir el tiempo de ejecución de un algoritmo de ordenamiento
def medir_tiempo_ordenamiento(algoritmo, lista):
    inicio = time.time()
    algoritmo(lista.copy())  # Copiamos la lista para no modificar la original
    fin = time.time()
    return fin - inicio

# Tamaños de las colecciones de datos
tamaños = [10, 100, 1000, 10000]

# Diccionario para almacenar los tiempos de cada algoritmo
tiempos = {"quickSort Sort": []}

for tamaño in tamaños:
    # Generar una lista aleatoria de tamaño 'tamaño'
    lista_aleatoria = random.sample(range(1, tamaño * 10), tamaño)

    # Medir el tiempo de ordenamiento para Insertion Sort
    tiempo = medir_tiempo_ordenamiento(quickSort, lista_aleatoria)
    tiempos["quickSort Sort"].append(tiempo)

# Generar la gráfica
for algoritmo, tiempo_lista in tiempos.items():
    plt.plot(tamaños, tiempo_lista, label=algoritmo)

plt.xlabel("Número de Datos")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs. Número de Datos para Algoritmos de Ordenamiento")
plt.legend()
plt.show()
```

```python
def quickSort(lista):
  # si la lista es de tamano 1 o menor se devuelve
  if len(lista) < 2:
    return lista
  # en otro caso se parte la lista y se resuelve recursivamente
  else:
    # se parte la lista original (n operaciones)
    menores, pivote, mayores = particion(lista)
    # 2*T(n/2)
    return quickSort(menores)+[pivote]+quickSort(mayores)

# Definicion del algoritmo para partir una lista en mayores y menores
def particion(lista):
  mayores, menores = [], []
  # pivote para realizar la comparación
  pivote = lista[0]
  for i in range(1,len(lista)):
    if lista[i] < pivote:
      menores.append(lista[i])
    else:
      mayores.append(lista[i])
  # se devuelven ambas listas y el pivote
  return menores, pivote, mayores
```
